```python
import numpy as np
import pandas as pd
import seaborn as sb
import os
```

##### Ubahlah dataset.csv sesuai dengan nama dataset kalian

```python
dataframe = pd.read_csv('glcm_result.csv')
dataframe.head()
```

```python
# os.listdir(
#     'segmented-dataset'
# )
```

##### Ubahlah label sesuai dengan label kalian

```python
clean_data  = dataframe.drop(columns=['Unnamed: 0', 'path'])
clean_data  = clean_data.replace ({
    'ayam': 0,
    'kalkun': 1,
    'puyuh': 2
})
clean_data.head()
```

```python
len(clean_data.columns)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix


x = clean_data.iloc[:,:-1]
y = clean_data.iloc[:,-1:]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

scaler = MinMaxScaler()
x_train_scalled = scaler.fit_transform(x_train)
x_test_scalled = scaler.transform(x_test)
```

##### Cek dokumentasi parameter pemanggilan fungsi pada list classifiers

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
names = ["K-Nearest Neighbors", 
         "Linear SVM"]

classifiers = [
    KNeighborsClassifier(5),
    SVC(gamma=2, C=1),
    ]
```

#### Lakukanlah analisis hasil prediksi

```python
for name, model in zip(names, classifiers):
  model.fit(x_train_scalled, y_train.values.ravel())
  y_pred_model = model.predict(x_test_scalled)
  print(f'## MODEL NAME : {name}')
  print(confusion_matrix(y_test, y_pred_model))
  print(classification_report(y_test, y_pred_model))
  print('\n')
```

```python
correlation = {}
label = clean_data['labels']
for col in clean_data.columns:
  correlation[col] = label.corr(clean_data[col])
sorted_correlation = sorted(correlation.items(), key=lambda x: x[1])
print(sorted_correlation)
len(sorted_correlation)
```

```python
positive_corr = {}
negative_corr = {}
for (k,v) in sorted_correlation:
  if(v>0):
    positive_corr[k] = v
  else:
    negative_corr[k] = v
len(positive_corr)
```

### PERBANDINGAN
Dalam membandingkan kedua model, K-Nearest Neighbors (KNN) dan Linear SVM, terdapat perbedaan sebagai berikut:

Akurasi: Akurasi berfungsi untuk mengukur sejauh mana model dapat memprediksi dengan benar. Pada model KNN, akurasi sebesar 0.60, sedangkan pada model Linear SVM, akurasi mencapai 0.73. Oleh karena itu, Linear SVM memiliki akurasi yang lebih tinggi daripada KNN.

Presisi dan Recall: Presisi mengukur sejauh mana hasil positif yang diprediksi oleh model adalah benar, sedangkan recall mengukur sejauh mana model dapat mengidentifikasi secara lengkap hasil positif yang sebenarnya. Model KNN memiliki presisi dan recall yang lebih rendah dibandingkan dengan model Linear SVM.

F1-Score: F1-score merupakan rata-rata antara presisi dan recall. Baik KNN maupun Linear SVM memiliki f1-score yang lumayan tinggi, tetapi model Linear SVM memiliki nilai f1-score yang sedikit lebih tinggi daripada KNN.


### KESIMPULAN
Berdasarkan perbandingan kedua model tersebut, dapat disimpulkan bahwa model Linear SVM memiliki hasil yang lebih baik daripada model K-Nearest Neighbors dalam memprediksi kelas pada dataset yang diberikan. Model Linear SVM memiliki akurasi yang lebih tinggi dan secara konsisten memberikan nilai presisi, recall, dan f1-score yang lebih baik untuk setiap kelas.
